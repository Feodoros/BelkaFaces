# Обзор MTCNN

```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import time
```

```python
test_image = cv2.cvtColor(cv2.imread('Images\\test.jpg'), cv2.COLOR_BGR2RGB) 
cage_image = cv2.cvtColor(cv2.imread('Images\\cage.jpg'), cv2.COLOR_BGR2RGB)
musk_image = cv2.cvtColor(cv2.imread('Images\\test.jpg'), cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1)
fig.set_size_inches(18.5, 10.5)

ax.set_title('test_image')
ax.imshow(test_image)
plt.show()
```

```python
def detect_faces(img):
    start_time = time.time()
    detector = MTCNN()
    
    faces = detector.detect_faces(img)
    work_time = round(time.time() - start_time, 3)
    return faces, work_time 
```

```python
def show_result(img=test_image):
    copy_img = img
    faces, work_time = detect_faces(img)
    coord_faces = list(map(lambda i: i[(list(faces[0].keys())[0])], faces))
    for (x, y, w, h) in coord_faces:
        cv2.rectangle(copy_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
    fig, ax = plt.subplots(1)
    fig.set_size_inches(18.5, 10.5)

    ax.set_title(str(work_time))
    ax.imshow(copy_img)
    plt.show()
        
```

```python
show_result()
```

```python
show_result(cage_image)
```
